# 01 — Policy ingestion (PDF → clean text)

**Purpose:** Load regulatory PDFs from `data/raw_policies/`, extract text, clean headers/footers/page numbers, preserve section headings. No scraping — PDFs are user-downloaded from official sources (see PHASE1_PROPOSAL.md).

**Data sources (traceable):** FDA (promotion, AE); EMA GVP; CDSCO PV; India Drugs and Magic Remedies Act 1954; ICH E2A/E2D; GDPR; HIPAA.

**Output:** One `.txt` per document in `data/processed/clean_text/`.

```python
import json
from pathlib import Path

# Paths relative to project root (run notebook from project root)
PROJECT_ROOT = Path.cwd() if Path.cwd().name != 'notebooks' else Path.cwd().parent
RAW_DIR = PROJECT_ROOT / 'data' / 'raw_policies'
CLEAN_DIR = PROJECT_ROOT / 'data' / 'processed' / 'clean_text'
CLEAN_DIR.mkdir(parents=True, exist_ok=True)

with open(RAW_DIR / 'sources_metadata.json') as f:
    sources = json.load(f)['sources']

print(f"Found {len(sources)} sources. PDFs expected in: {RAW_DIR}")
```

## PDF text extraction

Uses `pypdf` for extraction. For complex layouts, consider `pdfplumber` (already in requirements). We preserve structure by keeping newlines; cleaning step removes common header/footer and page-number patterns.

```python
import re

try:
    from pypdf import PdfReader
except ImportError:
    from PyPDF2 import PdfReader  # fallback


def extract_text_from_pdf(pdf_path: Path) -> list[tuple[int, str]]:
    """Extract text per page. Returns list of (page_num_1based, text)."""
    reader = PdfReader(str(pdf_path))
    pages = []
    for i, page in enumerate(reader.pages):
        text = page.extract_text() or ""
        pages.append((i + 1, text))
    return pages


def clean_page_text(text: str) -> str:
    """Remove common header/footer and page number patterns; collapse excess newlines."""
    # Remove standalone page numbers (e.g. " 12 " or "Page 12")
    text = re.sub(r'\bPage\s+\d+\b', '', text, flags=re.IGNORECASE)
    text = re.sub(r'^\s*\d+\s*$', '', text, flags=re.MULTILINE)
    # Remove repeated header/footer lines (same line repeated at top/bottom)
    lines = text.split('\n')
    if len(lines) > 4:
        # Often first 1–2 and last 1–2 lines are header/footer
        lines = lines[2:-2] if lines[0].strip() == lines[-1].strip() else lines
    text = '\n'.join(lines)
    # Collapse multiple newlines to at most two
    text = re.sub(r'\n{3,}', '\n\n', text)
    return text.strip()


def full_document_text(pages: list[tuple[int, str]], clean: bool = True) -> str:
    """Concatenate pages with optional cleaning."""
    parts = []
    for num, t in pages:
        if clean:
            t = clean_page_text(t)
        parts.append(t)
    return '\n\n'.join(parts)
```

```python
# Process each source: find PDF by 'file' name, extract, write .txt
results = []
for src in sources:
    pdf_name = src['file']
    pdf_path = RAW_DIR / pdf_name
    if not pdf_path.exists():
        print(f"Skip (not found): {pdf_name}")
        results.append({'file': pdf_name, 'status': 'not_found'})
        continue
    try:
        pages = extract_text_from_pdf(pdf_path)
        full_text = full_document_text(pages, clean=True)
        out_name = Path(pdf_name).stem + '.txt'
        out_path = CLEAN_DIR / out_name
        out_path.write_text(full_text, encoding='utf-8')
        results.append({'file': pdf_name, 'status': 'ok', 'pages': len(pages), 'out': out_name})
        print(f"OK: {pdf_name} → {out_name} ({len(pages)} pages)")
    except Exception as e:
        print(f"Error {pdf_name}: {e}")
        results.append({'file': pdf_name, 'status': 'error', 'error': str(e)})

print(f"\nDone. Output directory: {CLEAN_DIR}")
```
